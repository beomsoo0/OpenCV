```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((512,512,3), 255, np.uint8)
image = cv2.line(image, (0,0), (255,255), (255,0,0), 10)     # matplot은 RGB형식이므로 (255,0,0)은 빨간줄

plt.imshow(image)
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.rectangle(image, (20, 20), (255,255), (255,0,0), 3)

plt.imshow(image)
plt.show()

image = cv2.rectangle(image, (20, 20), (255,255), (255,0,0), -1)

plt.imshow(image)
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((512,512,3), 255, np.uint8)
image = cv2.circle(image, (255,255), 100, (0,255,0), 3)

plt.imshow(image)
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((512, 512, 3), 255, np.uint8)
point = np.array([[0,0], [500, 300], [300,500]])
image = cv2.polylines(image, [point], True, (0,0,255), 4)

plt.imshow(image)
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.putText(image, 'Hello', (100, 300), cv2.FONT_ITALIC, 2, (0, 0, 255))

plt.imshow(image)
plt.show()
```
