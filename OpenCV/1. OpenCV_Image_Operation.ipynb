```
import cv2
import matplotlib.pyplot as plt
import time

# for문 이용
start_time = time.time()
for i in range(100):
  for j in range(100):
    image[i, j] = [255,255,255]
print("==== %s seconds ====" %(time.time()-start_time))

# 슬라이싱 이용 (더 빠름)
start_time = time.time()
image[0:100, 0:100] = [0, 0, 0]
print("==== %s seconds ====" %(time.time()-start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

특정 부분 화소값 변경

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('beom.jpg')

# Numpy Slicing : ROI 처리 기능
roi = image[200:400, 250:450]

# ROI 단위로 이미지 복사하기
image[0:200, 0:200] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

ROI를 이용하여 특정 부분 추출

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('beom.jpg')
image[:, :, 2] = 0

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
